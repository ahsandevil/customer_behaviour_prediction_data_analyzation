```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')


df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))


df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)


labels = ['YOUNG', 'ADULT', 'MID', 'SENIOR']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)


frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

# 5. Drop the 'promo_code_used' column (as 'discount_applied' is similar)
df.drop('promo_code_used', axis=1, inplace=True)

print("Data cleaning and preprocessing steps completed successfully!")

# Display the first few rows of the processed DataFrame
display(df.head())
        
```
